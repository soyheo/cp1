```python
# 고객 세그멘테이션 by rfm
# Recency
df['event_time'] = pd.to_datetime(df['event_time'], format='%Y-%m-%d %H:%M:%S')
max_data = max(df['event_time'])
df['DateDiff'] = max_data - df['event_time']
rfm_r = df.groupby('user_id', as_index=False)['DateDiff'].min()
rfm_r.head()
```

```python
seconds_value = []

for i in range(len(rfm_r.DateDiff)):
    seconds_value.append(rfm_r.DateDiff[i].total_seconds())

rfm_r.DateDiff = seconds_value
```

```python
rfm_r.head()
```

```python
# Frequency
# 고객별 구매 빈도 계산
def get_purchase_count(series):
    res = len([ _ for x in series if x=='purchase'])
    return res

rfm_f = df.groupby('user_id').agg({'event_type':get_purchase_count}).reset_index()
rfm_f.sort_values('event_type', ascending=False)
```

```python
# Monetary
# 총 구매 합
rfm_m = df.groupby('user_id', as_index=False)['price'].sum()
rfm_m
```

```python
rfm_r = rfm_r.set_index('user_id')
rfm_f = rfm_f.set_index('user_id')
rfm_m = rfm_m.set_index('user_id')
rfm = rfm_r.join([rfm_f, rfm_m], how='inner')
rfm.columns = ['Recency', 'Frequency', 'Monetary']
```

```python
log_rfm = pd.concat([recency_log, frequency_log, monetary_log], axis=1)
log_rfm.head()
```

```python
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(log_rfm)
print('mean: ', rfm_scaled.mean(axis=0).round(2))
print('std: ', rfm_scaled.std(axis=0).round(2))
```

```python
# cluster
sse = {}

for k in range(1,10):
  kmeans = KMeans(n_clusters=k, random_state=1)
  kmeans.fit(rfm_scaled)
  sse[k] = kmeans.inertia_
# 결과 시각화
plt.title('The Elbow Method')
plt.xlabel('Values of k')
plt.ylabel('SSE')
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()
```

```python
# k = 4
kmeans = KMeans(n_clusters=4, max_iter=50, random_state=42)
kmeans.fit(rfm_scaled)

# 고객별 클러스터 지정
cluster_labels = kmeans.labels_
rfm_k4 = rfm.assign(Cluster = cluster_labels)
```

```python
rfm_k4.head()
```

```python
rfm_k4.groupby(['Cluster']).agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': ['mean','count']
}).round(0)
# Cluster0 : 가장 최근에, 자주, 많이 구매
# Cluster2 : 비교적 최근에, 두번째로 많이 구매
# Cluster3 : 비교적 예전에, 세번째로 많이 구매
# Cluster1 : 가장 예전에, 적게 구매 
```
